```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
t = np.linspace(0, 2*np.pi, 500)
a1, a2 = 1, 1  # amplitudes
phi0_values = [0, np.pi/2, np.pi]  # phase delays: in phase, quarter, out of phase
```

```python
fig, axes = plt.subplots(len(phi0_values), 1, figsize=(8, 6))

for i, phi0 in enumerate(phi0_values):
    # Two waves
    E1 = a1 * np.cos(t)
    E2 = a2 * np.cos(t - phi0)
    E_total = E1 + E2
    
    axes[i].plot(t, E1, label="Wave 1")
    axes[i].plot(t, E2, label="Wave 2")
    axes[i].plot(t, E_total, label="Superposition", linewidth=2, color="black")
    axes[i].set_title(f"Phase Delay φ₀ = {phi0:.2f} rad")
    axes[i].legend(loc="upper right")
    axes[i].set_ylim(-2.5, 2.5)
    axes[i].set_xlabel("Time (arbitrary units)")
    axes[i].set_ylabel("Amplitude")

plt.tight_layout()
plt.show()
```
